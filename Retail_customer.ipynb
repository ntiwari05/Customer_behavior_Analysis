```python
def dataset_overview(df):
    """
    Perform basic exploratory checks on a pandas DataFrame.

    Parameters:
        df (pd.DataFrame): Input dataset

    Returns:
        dict: Summary of dataset characteristics
    """

    overview = {}

    # Shape and size
    overview["shape"] = df.shape
    overview["size"] = df.size

    #Info
    overview["info"] = df.info()

    # Statistical Description
    overview['description'] = df.describe()

    # Missing values
    overview["missing_values"] = df.isnull().sum()

    # Duplicate rows
    overview["duplicate_rows"] = df.duplicated().sum()

    # Unique values per column
    overview["unique_values"] = df.nunique()

    return overview
```

```python
summary = dataset_overview(data)

print("Shape:", summary["shape"])
print("Size:", summary["size"])
print("\nMissing Values:\n", summary["missing_values"])
print("\nDuplicate Rows:", summary["duplicate_rows"])
print("\nUnique Values:\n", summary["unique_values"])
print("\nStatistical Description:\n", summary["description"])
print("\nInfo:\n", summary["info"])
```

```python
def unique_data_list(columns):
  for i in columns:
    print('\n')
    print(f'Unique values in {i} are')
    print(data[i].unique())
```

```python
print(f'Unique values  in each columns are/n',unique_data_list(['Gender','Category',
       'Size',  'Season', 'Subscription Status', 'Shipping Type','Discount Applied', 'Promo Code Used',
       'Payment Method', 'Frequency of Purchases']))
```

```python
# We can calculate median of different category using
#    data[data['Category']=='Clothing']['Review Rating'].median()
data['Review Rating']=data.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Now Preprocess columns name

data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data = data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# Creating a column purchase_frequency_days
frequency_map_days={
    'Fortnightly':14,
    'Monthly':30,
    'Quarterly':90,
    'Annually':365,
    'Weekly':7,
    'Bi-Weekly':14,
    'Every 3 Months' : 90
}

data['purchase_frequency_days'] = data['frequency_of_purchases'].map(frequency_map_days)
```
